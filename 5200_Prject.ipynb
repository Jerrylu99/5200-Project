```python
import pandas as pd

file_path = "/content/project/data/raw/public.dat"

# 老师示例（minimal ingestion）
col_specs = [(0,3),(4,6),(7,10)]
col_names = ["STORE_ID","STATE","EMP_WAVE1"]

df = pd.read_fwf(file_path,
                 colspecs=col_specs,
                 names=col_names)

df.head()
```

```python
df.to_csv("/content/project/data/processed/public_minimal_processed.csv",
          index=False)

print("saved.")
```
